```python
from unsloth import FastLanguageModel
import torch

max_seq_length = 2048
dtype = None
load_in_4bit = True

# model_name="unsloth/Qwen3-4B-Thinking-2507",
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name="unsloth/Qwen3-4B-Thinking-2507",
    dtype=dtype,
    max_seq_length=max_seq_length,
    load_in_4bit=load_in_4bit,
)

prompt_style = """ä»¥ä¸‹æ˜¯æè¿°ä»»åŠ¡çš„æŒ‡ä»¤ï¼Œä»¥åŠæä¾›è¿›ä¸€æ­¥ä¸Šä¸‹æ–‡çš„è¾“å…¥ã€‚
è¯·å†™å‡ºä¸€ä¸ªé€‚å½“å®Œæˆè¯·æ±‚çš„å›ç­”ã€‚
åœ¨å›ç­”ä¹‹å‰ï¼Œè¯·ä»”ç»†æ€è€ƒé—®é¢˜ï¼Œå¹¶åˆ›å»ºä¸€ä¸ªé€»è¾‘è¿è´¯çš„æ€è€ƒè¿‡ç¨‹ï¼Œä»¥ç¡®ä¿å›ç­”å‡†ç¡®æ— è¯¯ã€‚

### æŒ‡ä»¤ï¼š
ä½ æ˜¯ä¸€ä½ç²¾é€šåŒ»å­¦çŸ¥è¯†çš„åŒ»ç”Ÿï¼Œèƒ½å¤Ÿå›ç­”å…³äºç–¾ç—…ã€æ²»ç–—æ–¹æ¡ˆå’Œå¥åº·å»ºè®®çš„é—®é¢˜ã€‚
è¯·å›ç­”ä»¥ä¸‹åŒ»ç–—é—®é¢˜ã€‚

### é—®é¢˜ï¼š
{}

### å›ç­”ï¼š
<think>{}"""

question = "æˆ‘æœ€è¿‘æ€»æ˜¯æ„Ÿåˆ°ç–²åŠ³ï¼Œå¯èƒ½æ˜¯ä»€ä¹ˆåŸå› ï¼Ÿ"

FastLanguageModel.for_inference(model)

inputs = tokenizer(prompt_style.format(question, ""), return_tensors="pt").to('cuda')

outputs = model.generate(
    input_ids=inputs.input_ids,
    attention_mask=inputs.attention_mask,
    max_new_tokens=1200,
    use_cache=True,
)

response = tokenizer.batch_decode(outputs)
print(response[0])
```

```python
# å®šä¹‰ç»“æŸæ ‡è®°ï¼ˆEOS_TOKENï¼‰ï¼Œç”¨äºæŒ‡ç¤ºæ–‡æœ¬çš„ç»“æŸ
EOS_TOKEN = tokenizer.eos_token  # å¿…é¡»æ·»åŠ ç»“æŸæ ‡è®°

# å¯¼å…¥æ•°æ®é›†åŠ è½½å‡½æ•°
from datasets import load_dataset
# åŠ è½½æŒ‡å®šçš„æ•°æ®é›†ï¼Œé€‰æ‹©ä¸­æ–‡è¯­è¨€å’Œè®­ç»ƒé›†çš„å‰500æ¡è®°å½•
dataset = load_dataset("shibing624/medical", 'finetune', split = "train[0:200]", trust_remote_code=True)
# æ‰“å°æ•°æ®é›†çš„åˆ—åï¼ŒæŸ¥çœ‹æ•°æ®é›†ä¸­æœ‰å“ªäº›å­—æ®µ
print(dataset.column_names)
```

```python
train_prompt_style = """ä»¥ä¸‹æ˜¯æè¿°ä»»åŠ¡çš„æŒ‡ä»¤ï¼Œä»¥åŠæä¾›è¿›ä¸€æ­¥ä¸Šä¸‹æ–‡çš„è¾“å…¥ã€‚
è¯·å†™å‡ºä¸€ä¸ªé€‚å½“å®Œæˆè¯·æ±‚çš„å›ç­”ã€‚
åœ¨å›ç­”ä¹‹å‰ï¼Œè¯·ä»”ç»†æ€è€ƒé—®é¢˜ï¼Œå¹¶åˆ›å»ºä¸€ä¸ªé€»è¾‘è¿è´¯çš„æ€è€ƒè¿‡ç¨‹ï¼Œä»¥ç¡®ä¿å›ç­”å‡†ç¡®æ— è¯¯ã€‚

### æŒ‡ä»¤ï¼š
ä½ æ˜¯ä¸€ä½ç²¾é€šåŒ»å­¦çŸ¥è¯†çš„åŒ»ç”Ÿï¼Œèƒ½å¤Ÿå›ç­”å…³äºç–¾ç—…ã€æ²»ç–—æ–¹æ¡ˆå’Œå¥åº·å»ºè®®çš„é—®é¢˜ã€‚
è¯·å›ç­”ä»¥ä¸‹åŒ»ç–—é—®é¢˜ã€‚

### é—®é¢˜ï¼š
{}

### å›ç­”ï¼š
<æ€è€ƒ>
{}
</æ€è€ƒ>
{}"""

# å®šä¹‰ä¸€ä¸ªå‡½æ•°ï¼Œç”¨äºæ ¼å¼åŒ–æ•°æ®é›†ä¸­çš„æ¯æ¡è®°å½•
def formatting_prompts_func(examples):
    # ä»æ•°æ®é›†ä¸­æå–é—®é¢˜ã€å¤æ‚æ€è€ƒè¿‡ç¨‹å’Œå›ç­”
    inputs = examples["instruction"]
    cots = examples["input"]
    outputs = examples["output"]
    texts = []  # ç”¨äºå­˜å‚¨æ ¼å¼åŒ–åçš„æ–‡æœ¬
    # éå†æ¯ä¸ªé—®é¢˜ã€æ€è€ƒè¿‡ç¨‹å’Œå›ç­”ï¼Œè¿›è¡Œæ ¼å¼åŒ–
    for input, cot, output in zip(inputs, cots, outputs):
        # ä½¿ç”¨å­—ç¬¦ä¸²æ¨¡æ¿æ’å…¥æ•°æ®ï¼Œå¹¶åŠ ä¸Šç»“æŸæ ‡è®°
        text = train_prompt_style.format(input, cot, output) + EOS_TOKEN
        texts.append(text)  # å°†æ ¼å¼åŒ–åçš„æ–‡æœ¬æ·»åŠ åˆ°åˆ—è¡¨ä¸­
    return {
        "text": texts,  # è¿”å›åŒ…å«æ‰€æœ‰æ ¼å¼åŒ–æ–‡æœ¬çš„å­—å…¸
    }

dataset = dataset.map(formatting_prompts_func, batched = True)
dataset["text"][0]
```

```python
# å°†æ¨¡å‹åˆ‡æ¢åˆ°è®­ç»ƒæ¨¡å¼ã€‚
# FastLanguageModel.for_training æ˜¯ Unsloth æä¾›çš„æ–¹æ³•ï¼Œç¡®ä¿æ¨¡å‹å‡†å¤‡å¥½è¿›è¡Œå‚æ•°æ›´æ–°ï¼Œè€Œä¸æ˜¯ä»…ç”¨äºæ¨ç†ã€‚
FastLanguageModel.for_training(model)

# é…ç½®å¹¶è¿”å›ä¸€ä¸ªæ”¯æŒå‚æ•°é«˜æ•ˆå¾®è°ƒï¼ˆPEFTï¼‰çš„æ¨¡å‹ã€‚
# get_peft_model ä½¿ç”¨ LoRA æŠ€æœ¯ï¼Œåªæ›´æ–°æ¨¡å‹çš„éƒ¨åˆ†å‚æ•°ï¼Œä»è€Œå‡å°‘æ˜¾å­˜éœ€æ±‚å’Œè®¡ç®—å¼€é”€ã€‚
model = FastLanguageModel.get_peft_model(
    model,  # ä¼ å…¥ä¹‹å‰åŠ è½½çš„é¢„è®­ç»ƒæ¨¡å‹ï¼Œä½œä¸ºå¾®è°ƒçš„åŸºç¡€ã€‚
    r=16,   # è®¾ç½® LoRA çš„ç§©ï¼ˆrankï¼‰ï¼Œæ§åˆ¶æ–°å¢å¯è®­ç»ƒå‚æ•°çš„è§„æ¨¡ã€‚å€¼è¶Šå¤§ï¼Œæ¨¡å‹è°ƒæ•´èƒ½åŠ›è¶Šå¼ºï¼Œä½†æ˜¾å­˜éœ€æ±‚ä¹Ÿå¢åŠ ã€‚
    target_modules=["q_proj", "k_proj", "v_proj", "o_proj", "gate_proj", "up_proj", "down_proj"],
            # æŒ‡å®šéœ€è¦åº”ç”¨ LoRA çš„æ¨¡å‹æ¨¡å—ï¼Œè¿™äº›æ˜¯ Transformer æ¶æ„ä¸­çš„å…³é”®éƒ¨åˆ†ï¼ˆå¦‚æ³¨æ„åŠ›æœºåˆ¶å’Œå‰é¦ˆç½‘ç»œï¼‰ã€‚
    lora_alpha=16,
            # LoRA çš„ç¼©æ”¾å› å­ï¼Œå½±å“æ–°å¢å‚æ•°å¯¹æ¨¡å‹çš„è´¡çŒ®ç¨‹åº¦ã€‚é€šå¸¸ä¸ r æˆæ¯”ä¾‹è®¾ç½®ï¼Œè¿™é‡Œä¸º 16ã€‚
    lora_dropout=0,
            # è®¾ç½® LoRA å±‚çš„ dropout æ¯”ç‡ï¼Œç”¨äºé˜²æ­¢è¿‡æ‹Ÿåˆã€‚è¿™é‡Œè®¾ä¸º 0ï¼Œè¡¨ç¤ºä¸ä¸¢å¼ƒä»»ä½•å‚æ•°ã€‚
    bias="none",
            # æŒ‡å®šæ˜¯å¦ä¸º LoRA å‚æ•°æ·»åŠ åç½®é¡¹ã€‚"none" è¡¨ç¤ºä¸æ·»åŠ ï¼Œä¿æŒè½»é‡åŒ–ã€‚
    use_gradient_checkpointing="unsloth",
            # å¯ç”¨æ¢¯åº¦æ£€æŸ¥ç‚¹æŠ€æœ¯ï¼ŒUnsloth ä¼˜åŒ–ç‰ˆæœ¬èƒ½èŠ‚çœæ˜¾å­˜ï¼Œæ”¯æŒæ›´å¤§çš„æ‰¹é‡å¤§å°æˆ–æ¨¡å‹ã€‚
    random_state=3407,
            # è®¾ç½®éšæœºç§å­ï¼Œç¡®ä¿æ¯æ¬¡è¿è¡Œæ—¶æ¨¡å‹åˆå§‹åŒ–çš„éšæœºæ€§ä¸€è‡´ï¼Œä¾¿äºç»“æœå¤ç°ã€‚
    use_rslora=False,
            # æ˜¯å¦ä½¿ç”¨ Rank-Stabilized LoRAï¼ˆä¸€ç§æ”¹è¿›çš„ LoRA å˜ä½“ï¼‰ã€‚è¿™é‡Œè®¾ä¸º Falseï¼Œä½¿ç”¨æ ‡å‡† LoRAã€‚
    loftq_config=None,
            # è®¾ç½®æ˜¯å¦ä½¿ç”¨ LoftQï¼ˆä¸€ç§é‡åŒ–æŠ€æœ¯ï¼‰ã€‚è¿™é‡Œè®¾ä¸º Noneï¼Œè¡¨ç¤ºä¸å¯ç”¨ã€‚
)
# å¯¼å…¥ SFTTrainer ç±»ï¼Œç”¨äºç›‘ç£å¾®è°ƒè®­ç»ƒã€‚
# SFTTrainer æ˜¯ TRLï¼ˆTransformers Reinforcement Learningï¼‰åº“æä¾›çš„å·¥å…·ï¼Œé€‚åˆåŸºäºæŒ‡ä»¤çš„æ•°æ®é›†å¾®è°ƒæ¨¡å‹ã€‚
from trl import SFTTrainer

# å¯¼å…¥ TrainingArguments ç±»ï¼Œç”¨äºå®šä¹‰è®­ç»ƒçš„è¶…å‚æ•°ã€‚
# è¿™æ˜¯ Hugging Face Transformers åº“çš„æ ¸å¿ƒç±»ï¼Œå…è®¸çµæ´»é…ç½®è®­ç»ƒè¿‡ç¨‹ã€‚
from transformers import TrainingArguments

# å¯¼å…¥ Unsloth æä¾›çš„å‡½æ•°ï¼Œç”¨äºæ£€æŸ¥ç¡¬ä»¶æ˜¯å¦æ”¯æŒ bfloat16 æ•°æ®ç±»å‹ã€‚
# bfloat16 æ˜¯ä¸€ç§é«˜æ•ˆçš„åŠç²¾åº¦æ ¼å¼ï¼Œèƒ½åœ¨æ”¯æŒçš„ç¡¬ä»¶ä¸ŠåŠ é€Ÿè®­ç»ƒã€‚
from unsloth import is_bfloat16_supported

# åˆ›å»º SFTTrainer å®ä¾‹ï¼Œé…ç½®è®­ç»ƒæ‰€éœ€çš„æ¨¡å‹ã€æ•°æ®å’Œå‚æ•°ã€‚
trainer = SFTTrainer(
    model=model,                    # ä¼ å…¥ä¹‹å‰é…ç½®å¥½çš„æ¨¡å‹ï¼ˆå·²å¯ç”¨ LoRAï¼‰ã€‚
    tokenizer=tokenizer,            # ä¼ å…¥ä¸æ¨¡å‹é…å¥—çš„åˆ†è¯å™¨ï¼Œç”¨äºå¤„ç†æ–‡æœ¬æ•°æ®ã€‚
    train_dataset=dataset,          # ä¼ å…¥è®­ç»ƒæ•°æ®é›†ï¼ˆå·²æ ¼å¼åŒ–ä¸ºåŒ…å« "text" å­—æ®µï¼‰ã€‚
    dataset_text_field="text",      # æŒ‡å®šæ•°æ®é›†ä¸­åŒ…å«è®­ç»ƒæ–‡æœ¬çš„å­—æ®µåï¼Œè¿™é‡Œæ˜¯ "text"ã€‚
    max_seq_length=max_seq_length,  # è®¾ç½®æœ€å¤§åºåˆ—é•¿åº¦ï¼Œä¸æ¨¡å‹åŠ è½½æ—¶ä¿æŒä¸€è‡´ï¼ˆå¦‚ 2048ï¼‰ã€‚
    dataset_num_proc=2,             # è®¾ç½®æ•°æ®å¤„ç†çš„å¹¶è¡Œè¿›ç¨‹æ•°ï¼ŒåŠ é€Ÿæ•°æ®é¢„å¤„ç†ã€‚
    packing=False,                  # æ˜¯å¦å¯ç”¨åºåˆ—æ‰“åŒ…ã€‚False è¡¨ç¤ºä¸æ‰“åŒ…ï¼Œæ¯æ¡æ•°æ®ç‹¬ç«‹å¤„ç†ã€‚
    args=TrainingArguments(         # å®šä¹‰è®­ç»ƒè¶…å‚æ•°çš„é…ç½®å¯¹è±¡ã€‚
        per_device_train_batch_size=2,  # æ¯ä¸ªè®¾å¤‡ï¼ˆGPUï¼‰çš„æ‰¹æ¬¡å¤§å°ï¼Œè®¾ä¸º 2 ä»¥é€‚åº”æ˜¾å­˜é™åˆ¶ã€‚
        gradient_accumulation_steps=4,  # æ¢¯åº¦ç´¯ç§¯æ­¥æ•°ï¼Œç´¯ç§¯ 4 æ¬¡å°æ‰¹æ¬¡åæ›´æ–°å‚æ•°ï¼Œæ¨¡æ‹Ÿå¤§æ‰¹é‡è®­ç»ƒã€‚
        warmup_steps=5,                 # é¢„çƒ­æ­¥æ•°ï¼Œå­¦ä¹ ç‡åœ¨å‰ 5 æ­¥é€æ¸å¢åŠ ï¼Œç¨³å®šè®­ç»ƒã€‚
        max_steps=75,                   # æœ€å¤§è®­ç»ƒæ­¥æ•°ï¼Œæ§åˆ¶è®­ç»ƒæ—¶é•¿ï¼ˆæ­¥æ•° = æ•°æ®é‡ / æ‰¹æ¬¡å¤§å°ï¼‰ã€‚
        learning_rate=2e-4,             # å­¦ä¹ ç‡ï¼Œè®¾ä¸º 0.0002ï¼Œæ§åˆ¶å‚æ•°æ›´æ–°é€Ÿåº¦ã€‚
        fp16=not is_bfloat16_supported(),  # å¦‚æœä¸æ”¯æŒ bfloat16ï¼Œåˆ™ä½¿ç”¨ fp16ï¼ˆ16 ä½æµ®ç‚¹æ•°ï¼‰åŠ é€Ÿè®­ç»ƒã€‚
        bf16=is_bfloat16_supported(),      # å¦‚æœç¡¬ä»¶æ”¯æŒ bfloat16ï¼Œåˆ™å¯ç”¨å®ƒï¼Œå…¼é¡¾ç²¾åº¦å’Œé€Ÿåº¦ã€‚
        logging_steps=1,                   # æ¯ 1 æ­¥è®°å½•ä¸€æ¬¡è®­ç»ƒæ—¥å¿—ï¼Œæ–¹ä¾¿ç›‘æ§æŸå¤±å˜åŒ–ã€‚
        optim="adamw_8bit",                # ä½¿ç”¨ 8 ä½ AdamW ä¼˜åŒ–å™¨ï¼ŒèŠ‚çœæ˜¾å­˜å¹¶ä¿æŒæ€§èƒ½ã€‚
        weight_decay=0.01,                 # æƒé‡è¡°å‡ç³»æ•°ï¼Œè®¾ä¸º 0.01ï¼Œé˜²æ­¢æ¨¡å‹è¿‡æ‹Ÿåˆã€‚
        lr_scheduler_type="linear",        # å­¦ä¹ ç‡è°ƒåº¦å™¨ç±»å‹ï¼Œ"linear" è¡¨ç¤ºçº¿æ€§è¡°å‡ã€‚
        seed=3407,                         # éšæœºç§å­ï¼Œç¡®ä¿è®­ç»ƒç»“æœå¯å¤ç°ã€‚
        output_dir="outputs",              # è®­ç»ƒç»“æœï¼ˆå¦‚æ£€æŸ¥ç‚¹ï¼‰ä¿å­˜çš„ç›®å½•ã€‚
        report_to="none",                  # ä¸å°†è®­ç»ƒæ—¥å¿—æŠ¥å‘Šåˆ°å¤–éƒ¨å·¥å…·ï¼ˆå¦‚ WandBï¼‰ï¼Œä»…æœ¬åœ°è®°å½•ã€‚
    ),
)

# å¼€å§‹è®­ç»ƒæ¨¡å‹ã€‚
# trainer.train() æ‰§è¡Œå¾®è°ƒè¿‡ç¨‹ï¼Œæ ¹æ®é…ç½®æ›´æ–°æ¨¡å‹å‚æ•°ï¼Œå®Œæˆåä¿å­˜åˆ° output_dirã€‚
trainer.train()
```

```python
print(question) # æ‰“å°å‰é¢çš„é—®é¢˜
# å°†æ¨¡å‹åˆ‡æ¢åˆ°æ¨ç†æ¨¡å¼ï¼Œå‡†å¤‡å›ç­”é—®é¢˜
FastLanguageModel.for_inference(model)

# å°†é—®é¢˜è½¬æ¢æˆæ¨¡å‹èƒ½ç†è§£çš„æ ¼å¼ï¼Œå¹¶å‘é€åˆ° GPU ä¸Š
inputs = tokenizer([prompt_style.format(question, "")], return_tensors="pt").to("cuda")

# è®©æ¨¡å‹æ ¹æ®é—®é¢˜ç”Ÿæˆå›ç­”ï¼Œæœ€å¤šç”Ÿæˆ 4000 ä¸ªæ–°è¯
outputs = model.generate(
    input_ids=inputs.input_ids,  # è¾“å…¥çš„æ•°å­—åºåˆ—
    attention_mask=inputs.attention_mask,  # æ³¨æ„åŠ›é®ç½©ï¼Œå¸®åŠ©æ¨¡å‹ç†è§£å“ªäº›éƒ¨åˆ†é‡è¦
    max_new_tokens=4000,  # æœ€å¤šç”Ÿæˆ 4000 ä¸ªæ–°è¯
    use_cache=True,  # ä½¿ç”¨ç¼“å­˜åŠ é€Ÿç”Ÿæˆ
)

# å°†ç”Ÿæˆçš„å›ç­”ä»æ•°å­—è½¬æ¢å›æ–‡å­—
response = tokenizer.batch_decode(outputs)

# æ‰“å°å›ç­”
print(response[0])
```
